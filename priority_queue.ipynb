## Introduction to Priority Queue

**Priority Queue**:
> A Priority Queue is a special type of queue where elements are removed based on priority, not by arrival time (FIFO).A priority queue does NOT follow FIFO.
Instead, the element with the highest priority gets removed first, not the one that came first.

High priority â†’ removed first
Low priority â†’ removed later

### Real Life Example
1. **Hospital Emergency Room (BEST example)**
Patients are treated based on priority, not who came first:

Even if Ali came first, Sara is treated first. This is a perfect priority queue.

### implementing a Priority Queue
Priority Queue is a concept, not a specific structure. So it can be implemented using different underlying data structures. But each implementation has different time complexity, performance, and use cases.

#### Different Ways to Implement a Priority Queue
 **1. Using Array (Unsorted Array)**

Insertion: O(1)
Deletion: O(n) â†’ you must search the max/min priority

âœ” Easy to implement
âœ˜ Slow when deleting max/min

**2. Using Sorted Array**

Insertion: O(n) â†’ insert at correct position
Deletion: O(1) â†’ remove last or first element (already sorted)

âœ” Fast deletion
âœ˜ Slow insertion

**3ï¸. Using Linked List**

Same as array logic but flexible memory.
Unsorted LL: Insert O(1), delete O(n)
Sorted LL: Insert O(n), delete O(1)

âœ” Good when insert/delete patterns matter
âœ˜ Still not optimal

**4. Binary Search Tree (BST / Balanced BST)**

Insertion: O(log n)
Deletion: O(log n)

âœ” Good performance
âœ˜ Worst case becomes O(n) if BST becomes unbalanced

(Using AVL/Red-Black Tree fixes this.)

**5. Using Heap (Best Implementation)â­**

This is the most efficient and commonly used structure for priority queues.

Why?

A Binary Heap gives:
Insertion: O(log n)
Deletion (extract-max/min): O(log n)

Find max/min: O(1)

âœ” Fast
âœ” Memory efficient
âœ” Used in real systems
âœ” What Pythonâ€™s heapq uses

ðŸ‘‰ This is why heaps are the standard implementation of priority queues.

### Priority Queue with Only Python List


```python
# -----------------------------------------------------------Hospital Emergency Queue 
queue = []

queue.append(("Normal Fever", 1))
queue.append(("Broken Arm", 3))
queue.append(("Chest Pain", 5))  # highest priority
queue.append(("Flu", 2))

# remove highest priority patient
highest = -1
pos = -1

for i in range(len(queue)):
    if queue[i][1] > highest:
        highest = queue[i][1]
        pos = i

patient = queue.pop(pos)

print("Treated:", patient)
print("Remaining Patients:", queue)
```

```python
#-----------------------------------------------Real Life: Airport Runway Scheduling
runway = []

runway.append(("Passenger Plane", 1))
runway.append(("Cargo Plane", 2))
runway.append(("Emergency Plane", 5))

highest = -1
idx = -1

for i in range(len(runway)):
    if runway[i][1] > highest:
        highest = runway[i][1]
        idx = i

first_land = runway.pop(idx)
print("First to land:", first_land)
```
