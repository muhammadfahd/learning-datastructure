## Function Call (Concept First)

* **What is a Function?** A function is a named block of code that performs a specific task and can be reused.
* **What is a Function Call?** When we use a function by its name, we say we are calling the function.
* **Real-life example**
Think of a mobile contact:
You save a contact as Ali
When you tap Ali, the phone executes all the steps behind that contact

```python
'''
When greet() is called:
Control jumps inside the function
Code executes
Control comes back

'''
def greet():
    print("Hello, World!")

greet()   # function call
```

### Function Calling Another Function

```python
''''
Functions can call other functions

Python remembers where it came from
'''
def add():
    print("Addition function")

def main():
    print("Main function start")
    add()
    print("Main function end")

main()
```

### unction Call Stack (Very Simple Explanation)
Python uses a stack to remember function calls.
 Like a stack of plates:
Last function called  executed first


```python
def fun1():
    print("Inside fun1")

def fun2():
    fun1()
    print("Inside fun2")

fun2()

#order of execution fun2()  fun1()  back to fun2()
```

## ntroduction to Recursion
**What is Recursion?**
Recursion is when a function calls itself. A problem solved by breaking it into smaller versions of the same problem.

1. **Two Important Parts of Recursion**
* **Base Case**  when recursion stops
* **Recursive Case**  function calls itself

2. Without base case  infinite loop

```python
# First Recursive Example (Counting)
def count(n):
    if n == 0:        # base case
        return
    print(n)
    count(n - 1)     # recursive call
```

```python
count(50)
```

```python
# Classic Beginner Recursion Example - Factorial of Number
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

```python
factorial(0)
```

```python
# sum of number
def sum_n(n):
    if n == 0:
        return 0
    return n + sum_n(n-1)

print(sum_n(4))
```
