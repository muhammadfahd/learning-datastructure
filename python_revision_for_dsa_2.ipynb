## 2. Tuples
A tuple is a collection which is ordered and unchangeable. Tuples are written with round brackets.
> thistuple = ("apple", "banana", "cherry")
- Ordered
- Immutable
- Useful for fixed data

```python
person = ("Ali", 21, "Student")
print(person)
```

### Tuples methods
- **Access** like list using indexing
- **updating value** Once a tuple is created, you cannot change its values. Tuples are unchangeable, or immutable as it also is called. You can convert the tuple into a list, change the list, and convert the list back into a tuple.
- can loop same like list for example for loop,while loop, range or len function


## 3. Sets
> Set is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.
- Unordered
- Unique values only
- Fast membership testing

```python
set1 = {"apple", "banana", "cherry"}
set2 = {1, 5, 7, 9, 3}
set3 = {True, False, False}
```

```python
unique_nums = {1,2,3,3,4}
print(unique_nums)
print(2 in unique_nums)
```

### Adding Item  & Removing Item

## 4. Dictionaries
Dictionaries are used to store data values in key:value pairs. A dictionary is a collection which is ordered*, changeable and do not allow duplicates.
- Key-value pairs
- Very fast lookups
- Useful for mapping relationships
- The values in dictionary items can be of any data type:

```python
student = {"name":"Fahad","age":22,"city":"Rawalpindi"}
print(student["name"])
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict["brand"])
```

### Accessing, Changing Items
- You can access the items of a dictionary by referring to its key name, inside square brackets:
- You can change the value of a specific item by referring to its key name:
- Adding an item to the dictionary is done by using a new index key and assigning a value to it:


```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict["model"]
print(x)
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["year"] = 2018
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["color"] = "red"
print(thisdict)
```

### 1. Make a list of marks and find highest.


```python
# 1. Make a list of marks and find highest.
marks = [78, 92, 85, 67, 90, 88]

highest = max(marks)
print("Marks:", marks)
print("Highest mark:", highest)
```

### 2. Make a tuple storing (name, age, city).

```python
student_info = ("Fahad", 22, "Rawalpindi")

print("Student Tuple:", student_info)
print("Name:", student_info[0])
print("Age:", student_info[1])
print("City:", student_info[2])
```

### 3. Make a set of 10 numbers & remove duplicates.

```python
numbers = [1, 2, 3, 4, 5, 5, 6, 7, 7, 8]  # contains duplicates

unique_numbers = set(numbers)

print("Original list:", numbers)
print("Set with unique numbers:", unique_numbers)
```
